```python
import numpy as np
import pandas as pd
import re
```

```python
base = pd.read_excel(r'C:\Users\alex_\OneDrive\Área de Trabalho\EMERGENCIAL.xlsx')
```

```python
def numeros(x):
    texto = re.sub('[^0-9]', '', str(x))
    return texto
def maiuscula(x):
    up = str(x).upper()
    if up is not np.nan:  
        return up
def telefone(x):
    tel = re.sub('[^0-9]', '', str(x))
    if len(tel) == 11:
        return tel
def empresa(x):
    if x == 'ACRE': 
        return 30   
    elif x == 'MATO GROSSO':
        return 6
    elif x == 'MATO GROSSO DO SUL':
        return 10
    elif x == 'MINAS GERAIS':
        return 1
    elif x == 'NOVA FRIBURGO':
        return 2
    elif x == 'PARAIBA':
        return 5
    elif x == 'RONDÔNIA':
        return 20
    elif x == 'SÃO PAULO':
        return 9
    elif x == 'SERGIPE':
        return 3
    elif x == 'TOCANTINS':
        return 8
    else:
        return 999 
def canal(x):
    if x == 'CALLCENTER':
        return 2
    elif x == 'GISA':
        return 1
    else:
        999
def tipo_os(x):
    if x == 'CABO PARTIDO':
        return 'T'
    elif x == 'FALTA DE ENERGIA':
        return 'T'
    elif x == 'RELIGAÇÃO':
        return 'C'
    elif x == 'RISCO DE VIDA':
        return 'T'
    elif x == 'VARIAÇÃO DE TENSÃO':
        return 'T'
    else:
        'NI'
def codserv(x):
    if x == 'CABO PARTIDO':
        return 901
    elif x == 'FALTA DE ENERGIA':
        return 900
    elif x == 'RELIGAÇÃO':
        return 3
    elif x == 'RISCO DE VIDA':
        return 903
    elif x == 'VARIAÇÃO DE TENSÃO':
        return 902
    else:
        999
def sitios(x):
    if x == 'CANCELADA':
        return 'X'
    
    elif x == 'EXECUTADA':
        return 'T'
    elif x is np.nan:
        return ""
    else:
        return 'P'
def pto_controle(x):
    if x == 'CANCELADA':
        return 5
    elif x == 'DESPACHADA':
        return 1
    elif x == 'EM EXECUÇÃO':
        return 2
    elif x == 'EXECUTADA':
        return 3
    elif x == 'PENDENTE':
        return 0
def controle_os(x):
    if x == 5.0:
        return 'CANCELADA'
    elif x == 1.0:
        return 'DESPACHADA'
    elif x == 2.0:
        return 'EM EXECUCAO'
    elif x == 3.0:
        return 'EXECUTADA'
    elif x == 0.0:
        return 'ABERTA'
def situacao_os(x):
    if x != np.nan:
        if x == 'P':
            return 'PENDENTE'
        elif x == 'X':
            return 'CANCELADA'
        elif x == 'T':
            return 'TRANSMITIDA'
```

```python
oss = pd.DataFrame(
    columns = ['ID','CODEMP','PROTOCOLO','NUMCDC','NUMOS','NOME_CLIENTE','ENDERECO','BAIRRO','TELEFONE','LOCSERV','TIPOLOCAL',
               'NUMOS_ORIG','TIPO_OS','CODSERV','SERVICO','NUM_OBJETO','LOGIN_ABER','COD_CANAL','CANAL_ATENDIMENTO',
               'COD_LIGACAO','COD_ATENDIMENTO','CODMUN','NOMEMUN','NOMEREG','CODAGA_POLO','NOMEAGA_ATA','USUARIO_OS',
               'CODGRUPO','GRUPO_TENSAO','INDTIPO_INTERACAO','DESC_INTERACAO','IND_INICIATIVA_ABERTURA_OS',
               'COD_STATUS_LIGNOVA','STATUS_LIGA_NOVA','SITOS','SITUACAO_OS','PTO_CONTROLE','CONTROLE_OS','DSCIMPED',
               'CODMOTV','DSCMOTV','CODMOT_CAN','DSCMOT_CAN','OBS_ATENDENTE','SITOS_WEB','OBS_BAIXA','DT_INTERACAO','DATASOL',
               'DATAINITRAB','DATASAIDA','DATATERTRAB','DT_PREVISAO','DATADIG','DATATRANSF','DH_CRITICA','DTHULT_VRS_UC',
               'DTHALT','DTH_INICIO_CTBZC_ATDTO','DT_LIGACAO','QTD_OS_FILHAS','DT_CARGA','MEDIDOR','LEITURA','ABRANGENCIA',
               'CAUSA','DEFEITO_FALHA','ID_ELETRICISTA','MEDIDOR_RETIRADO','LEITURA_MED_RETIRADO','IND_CONCLUSAO',
               'PONTO_REFERENCIA'])
oss['CODEMP'] = base['Empresa'].apply(empresa)
oss['PROTOCOLO'] = base['ID única'].apply(numeros)
oss['NUMCDC'] = base['Numero da Unidade Consumidora'].apply(numeros)
oss['NOME_CLIENTE'] = base['Nome do Cliente'].apply(maiuscula)
oss['NUMOS'] = base['Criado em'].apply(numeros)+base['ID única'].apply(numeros)
oss['ENDERECO'] = base['Endereço'].apply(maiuscula)
oss['TELEFONE'] = np.where(base['Telefone - Celular'] is np.nan, base['Telefone - Trabalho'],base['Telefone - Celular'])
oss['TELEFONE'] = oss['TELEFONE'].apply(telefone)
oss['TIPO_OS'] = base['Tipo de Solicitação'].apply(tipo_os)
oss['CODSERV'] = base['Tipo de Solicitação'].apply(codserv)
oss['SERVICO'] = base['Tipo de Solicitação'].apply(maiuscula)
oss['LOGIN_ABER'] = base['LOGIN']
oss['COD_CANAL'] = base['Canal Abertura'].apply(canal)
oss['CANAL_ATENDIMENTO'] = base['Canal Abertura'].apply(maiuscula)
oss['SITOS'] = base['Status'].apply(sitios)
oss['SITUACAO_OS'] = base['Status']
oss['PTO_CONTROLE'] = oss['SITUACAO_OS'].apply(pto_controle)
oss['CONTROLE_OS'] = oss['PTO_CONTROLE'].apply(controle_os)
oss['OBS_ATENDENTE'] = base['Relato/Descrição'].apply(maiuscula)
oss['DT_INTERACAO'] = base['Criado em']
oss['DATASOL'] = base['Criado em']
oss['DATATERTRAB'] = base['Data Encerramento']
oss['ABRANGENCIA'] = base['Abrangência'].apply(numeros)
oss['PONTO_REFERENCIA'] = base['Ponto de referência'].apply(maiuscula)
oss.head()
```
